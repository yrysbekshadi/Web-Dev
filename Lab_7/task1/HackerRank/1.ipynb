```python
students = []
for i in range(int(input())):
    name = input()
    score = float(input())
    students.append([name, score])
second_lowest = sorted(set(score for i, score in students))[1]
print("\n".join(sorted(name for name, score in students if score == second_lowest)))
```

```python
n = int(input())
student_marks = {}
for _ in range(n):
    name, *scores = input().split()
    student_marks[name] = list(map(float, scores))

query_name = input()
print(f"{sum(student_marks[query_name]) / len(student_marks[query_name]):.2f}")
```

```python
n = int(input())
lst = []

for i in range(n):
    cmd, *args = input().split()
    if cmd == "insert":
        lst.insert(int(args[0]), int(args[1]))
    elif cmd == "print":
        print(lst)
    elif cmd == "remove":
        lst.remove(int(args[0]))
    elif cmd == "append":
        lst.append(int(args[0]))
    elif cmd == "sort":
        lst.sort()
    elif cmd == "pop":
        lst.pop()
    elif cmd == "reverse":
        lst.reverse()
```
